## The "Lang" Family: Four Products, One Team, Four Different Jobs
### 1 LangGraph — the foundation
LangGraph is a **low-level orchestration framework**. It gives you fine-grained control over how an application flows: branching logic,
loops, persistence (remembering state between steps), and durable execution (being able to pause and resume a long-running process).

### 2 LangChain
LangChain provides `create_agent` for building agents built *on top of* LangGraph, which means you get LangGraph's reliability (durable execution,
persistence, human-in-the-loop support).

> "LangChain's agents are built on top of LangGraph. This allows us to take advantage of
> LangGraph's durable execution, human-in-the-loop support, persistence, and more."

### 3 Deep Agents
Deep Agents is a layer built *on top of* LangChain's `create_agent`. Deep Agents gives you a harness that already comes with planning tools, a virtual filesystem, and the ability to spawn sub-agents — all pre-wired.

### 4 LangSmith
LangSmith is not something you "build an agent with." It's an **observability and evaluation platform** — it watches agents built with any of the other three tools (and, since it supports OpenTelemetry, technically almost anything else too) and lets you see exactly what happened during a run.

### LangSmith 
* LangSmith is an **observability and evaluation platform** — it watches agents built with any of the other three tools (and, since it supports
OpenTelemetry, technically almost anything else too) and lets you see exactly what happened during a run.

#### Problem Langsmith solve
* Agent decides what to do at *runtime*, based on a conversation it has with itself (calling tools, reasoning about results, calling more tools). 
* You cannot "read the code" to know what an agent actually did on a specific run, because the code only describes what it *could* do, not what it *did* do on that particular call.
* LangSmith solves this by recording a **trace** — a complete, record of every model call, tool call, and decision the agent made, in order. 
* When something goes wrong, you don't guess. You open the trace and read exactly what happened, step by step.

**Two environment variables turn this on:**
```
LANGSMITH_TRACING=true
LANGSMITH_API_KEY=your-key-here
```
**A note on cost:** LangSmith's free Developer tier includes 5,000 traces per month.

## Agent
> **An agent is a model calling tools in a loop until a given task is complete.**
> **A harness is everything around that loop.**

**The model** is the raw language model — GPT, Claude, Gemini, whatever you plug in. 
**The harness** is everything that turns that raw capability into something useful:
- The **system prompt** — instructions on how the agent should behave
- The **tools** — what it's actually allowed to reach for and use
- The **middleware** — checkpoints that shape its behavior at every step

### The three tiers, precisely
```
LangGraph         →  Build the harness from raw parts yourself. Maximum control, most effort.
LangChain         →  create_agent gives you a pre-configured, highly customizable harness.
Deep Agents       →  create_deep_agent gives you a harness that already includes planning, a filesystem, and sub-agent spawning, pre-wired.
```
