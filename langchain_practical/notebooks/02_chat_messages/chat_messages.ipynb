## Messages: System, Human, AI, and Tool
**Messages List** — a sequence of objects, where each one is explicitly labeled with who "said" it:

- **`SystemMessage`** — instructions for how the model should behave.
- **`HumanMessage`** — what the user actually said or asked.
- **`AIMessage`** — what the model has responded.
- **`ToolMessage`** — the result of a tool call, sent back to the model so it can continue reasoning with that new information.

```python

import os

from dotenv import load_dotenv
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI

load_dotenv()
```

```python
model = ChatOpenAI(model="gpt-5-nano")
```

### Start with system message and first question

```python
messages = [
    SystemMessage(content="You are a helpful coding tutor who gives clear, concise one liner explanations."),
    HumanMessage(content="What is Python?"),
]

print("👤 User: What is Python?")

# First exchange
response1 = model.invoke(messages)
print(f"\n🤖 AI: {response1.content}")
messages.append(AIMessage(content=str(response1.content)))
```

### Second exchange - AI remembers the context

```python
print("\n👤 User: Can you show me a simple example?")
messages.append(HumanMessage(content="Can you show me a simple example?"))

response2 = model.invoke(messages)
messages.append(AIMessage(content=str(response2.content)))
print(f"\n🤖 AI: {response2.content}")
```

### Third exchange - AI still remembers everything

```python
print("\n👤 User: What did you printed in your previous answer?")
messages.append(HumanMessage(content="What did you printed in your previous answer?"))

# the 3rd AI response is not added to conversation history since it is the last in the conversation
response3 = model.invoke(messages)
print(f"\n🤖 AI: {response3.content}")

print("\n\n✅ Notice how the AI maintains context throughout the conversation!")
print(f"📊 Total messages in history: {len(messages)} messages, that include 1 system message, 3 Human messages and 2 AI responses")
```

## Streaming Responses
### Streaming LLM Response using `.stream()` & `AIMessageChunk` object
* When you call `.invoke()`, you wait for the complete response and get back one `AIMessage`. 
* When you call `.stream()` instead, you get the response piece by piece, as it's generated and each piece is a distinct `AIMessageChunk` object
* The genuinely useful part: these chunks can be **added together** with the `+` operator, and the result is a fully reconstructed message.
* **Why streaming matters in practice:** it's what powers real-time chat interfaces, where text appears to the user as it's generated.

```python
chunks = []
full_message = None

for chunk in model.stream("Write one short sentence about the ocean."):
    chunks.append(chunk)
    print(repr(chunk.text), " <- chunk type:", type(chunk).__name__)
    full_message = chunk if full_message is None else full_message + chunk

print()
print("Number of chunks received:", len(chunks))
print("Reconstructed full message:", full_message.content)
print("Reconstructed type:", type(full_message).__name__)
```
